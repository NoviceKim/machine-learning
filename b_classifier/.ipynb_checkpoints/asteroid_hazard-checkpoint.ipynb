### Îç∞Ïù¥ÌÑ∞ ÌëúÏ§ÄÌôî

#### ÏÉÅÎåÄ ÏãúÏÜçÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Îç∞Ïù¥ÌÑ∞ ÌëúÏ§ÄÌôî(StandardScaler) ÌõÑ, Ïù¥ÏÉÅÏπò Í≤ÄÏÇ¨
- Ïù¥ featureÍ∞Ä Ï∂©Îèå ÌôïÎ•†Ïóê ÏòÅÌñ•ÏùÑ ÌÅ¨Í≤å ÎØ∏ÏπúÎã§Í≥† ÌåêÎã®ÌñàÏúºÎ©∞,  
  Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨ Ïó≠Ïãú ÌëúÏ§Ä Ï†ïÍ∑ú Î∂ÑÌè¨Ïóê Í∞ÄÏû• Í∞ÄÍπåÏõ†Í∏∞ ÎïåÎ¨∏

```python
from scipy import stats
import numpy as np

# ÏÉÅÎåÄ ÏãúÏÜçÏùò Î∂ÑÌè¨Î•º ÏÑ† Í∑∏ÎûòÌîÑ Ìè¨Ìï®Ìï¥ÏÑú ÏãúÍ∞ÅÌôî
kde = stats.gaussian_kde(ah_df['Relative Velocity'])
line = np.linspace(0, 160682, 2000, endpoint=True)

plt.hist(ah_df['Relative Velocity'], density=True, bins=100, alpha=0.5)
plt.plot(line, kde(line))
plt.show()
```

```python
# Ïù∏Îç±Ïä§ÏôÄ ÏÉÅÎåÄ ÏãúÏÜçÎßå Î∂ÑÎ¶¨Ìïú ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
rv_df = ah_df.loc[:, 'Relative Velocity'].reset_index()

rv_df
```

```python
from sklearn.preprocessing import StandardScaler

# ÏÉÅÎåÄ ÏãúÏÜç ÌëúÏ§ÄÌôî
std = StandardScaler()
std_rv = std.fit_transform(rv_df[['Relative Velocity']])

std_rv
```

```python
# ÌëúÏ§ÄÌôîÎêú Í∞íÏùÑ ÏõêÎûò Ïª¨ÎüºÏóê Ïû¨Ìï†Îãπ
rv_df['Relative Velocity'] = std_rv

rv_df
```

```python
l_cond = rv_df['Relative Velocity'] >= -1.96
u_cond = rv_df['Relative Velocity'] <= 1.96

cond = l_cond & u_cond

# ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ÏóêÏÑú ÏúÑ Ï°∞Í±¥ÏãùÏùÑ ÎßåÏ°±ÌïòÎäî Ïù∏Îç±Ïä§Îßå ÎÇ®ÍπÄÏúºÎ°úÏÑú Ïù¥ÏÉÅÏπò Ï†úÍ±∞
ah_df = ah_df.iloc[rv_df[cond].index].reset_index(drop=True)

ah_df
```

### Ïò§Ï∞® ÌñâÎ†¨
- ÏúÑ Î™®Îç∏Ïùò Ïò§Ï∞® ÌñâÎ†¨ÏùÑ ÏãúÍ∞ÅÌôî

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

# ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ÏôÄ ÏòàÏ∏° Í∞ùÏ≤¥Î•º Ï†ÑÎã¨Î∞õÎäîÎã§.
def get_evaluation_by_custom_proba(y_test, prediction, visualize=False):
#     Ïò§Ï∞® ÌñâÎ†¨
    confusion = confusion_matrix(y_test, prediction)
#     Ï†ïÌôïÎèÑ
    accuracy = accuracy_score(y_test , prediction)
#     Ï†ïÎ∞ÄÎèÑ
    precision = precision_score(y_test , prediction)
#     Ïû¨ÌòÑÏú®
    recall = recall_score(y_test , prediction)
#     F1 score
    f1 = f1_score(y_test, prediction)
#     ROC-AUC
    roc_auc = roc_auc_score(y_test, prediction)
    
    print('Ïò§Ï∞® ÌñâÎ†¨')
    print(confusion)
    print('Ï†ïÌôïÎèÑ: {0:.4f}, Ï†ïÎ∞ÄÎèÑ: {1:.4f}, Ïû¨ÌòÑÏú®: {2:.4f}, F1: {3:.4f}, ROC-AUC: {4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
    print("#" * 80)
    
    if visualize:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_predictions(y_test, prediction, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
            
        plt.tight_layout()
        plt.show()
```

```python
get_evaluation_by_custom_proba(y_test, prediction, visualize=True)
```

### Î≥∏ Î™®Îç∏Ïùò Í∏∞Ìöç Î∞©Ìñ•
- ÏòàÏ∏° Ïã§Ìå®Ïùò Í≤ΩÏö∞Ïùò ÏàòÎäî 2Í∞ÄÏßÄ
> 1. ÏúÑÌóòÏÑ±Ïù¥ ÏóÜÎã§Í≥† ÏòàÏ∏°Ìïú ÏÜåÌñâÏÑ±Ïù¥ Ïã§Ï†úÎ°úÎäî ÏúÑÌóòÌïú Í≤ΩÏö∞ (FN)
> 2. ÏúÑÌóòÌïòÎã§Í≥† ÏòàÏ∏°Ìïú ÏÜåÌñâÏÑ±Ïù¥ Ïã§Ï†úÎ°úÎäî ÏúÑÌóòÌïòÏßÄ ÏïäÏùÄ Í≤ΩÏö∞ (FP)

- Îëò Ï§ë Îçî Ïã¨Í∞ÅÌïú Í≤ÉÏùÄ Ï†ÑÏûêÏùò ÏÉÅÌô©Ïù¥Îãà, Ïù¥Î≤à Î™®Îç∏ÏóêÏÑúÎäî FNÏùò ÎπÑÏ§ëÏùÑ ÎÇÆÏ∂∞Ïïº ÌïúÎã§.
- Îî∞ÎùºÏÑú ÏûÑÍ≥ÑÍ∞í(Threshold Value)Î•º ÎÇÆÏ∂§ÏúºÎ°úÏÑú Ïû¨ÌòÑÏú®ÏùÑ Ìñ•ÏÉÅÏãúÏºúÏïº ÌïúÎã§.

### Precision(Ï†ïÎ∞ÄÎèÑ) / Recall(Ïû¨ÌòÑÏú®) trade-off

```python
# Îëò Ï§ë ÌôïÎ•†Ïù¥ ÎÜíÏùÄ Ï™ΩÍ≥º, Í∑∏Í≤å Ïñ¥Îäê Ï™ΩÏù∏ÏßÄÎ•º Ï∂úÎ†•
proba = dtc.predict_proba(X_test)
prediction = dtc.predict(X_test)

for proba, predict in zip(proba.tolist(), prediction.tolist()):
    print(proba, predict)
```

#### Custom proba

```python
from sklearn.preprocessing import Binarizer

# Í∏∞Î≥∏ ÏûÑÍ≥ÑÍ∞íÏùÑ 0.5Î°ú ÏÑ§Ï†ï
threshold = 0.5

proba = dtc.predict_proba(X_test)
prediction = dtc.predict(X_test)
```

```python
# ÏßÅÏ†ë proba Î∞è prediction ÏÉùÏÑ±
custom_proba = proba[:, 1].reshape(-1, 1)
binarizer = Binarizer(threshold=threshold).fit(custom_proba)
custom_prediction = binarizer.transform(custom_proba)

get_evaluation_by_custom_proba(y_test, custom_prediction, visualize=True)
```

```python
# ÌòÑÏû¨ custom_proba Í∞í Ï∂úÎ†•
print(custom_proba)
```

```python
# ÌòÑÏû¨ custom_prediction Í∞í Ï∂úÎ†• (T/F)
print(custom_prediction)
```

```python
# ÏûÑÍ≥ÑÍ∞íÏóê Îî∞Î•∏ Ïò§Ï∞® ÌñâÎ†¨ÏùÑ Íµ¨Ìï¥Ï£ºÎäî Ìï®Ïàò
def get_evaluation_by_thresholds(y_test, proba_class1, thresholds):
    for threshold in thresholds:
        print(threshold)
        binarizer = Binarizer(threshold=threshold).fit(proba_class1)
        custom_prediction = binarizer.transform(proba_class1)
        get_evaluation_by_custom_proba(y_test, custom_prediction)
```

> - ÏïÑÎûòÏùò Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìï¥ÏÑú Ìï¥Îãπ Î≥ÄÌôîÎ•º ÏãúÍ∞ÅÌôî

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Ï†ïÎ∞ÄÎèÑ - Ïû¨ÌòÑÏú® trade-offÎ•º ÏãúÍ∞ÅÌôîÌïòÎäî Ìï®Ïàò
def precision_recall_curve_plot(y_test , proba_class1):

    precisions, recalls, thresholds = precision_recall_curve(y_test, proba_class1)
    
    # XÏ∂ï: threshold
    # YÏ∂ï: Ï†ïÎ∞ÄÎèÑ, Ïû¨ÌòÑÏú® 
    # Ï†ïÎ∞ÄÎèÑÎäî Ï†êÏÑ†ÏúºÎ°ú ÌëúÏãú
    plt.figure(figsize=(8,6))
    threshold_boundary = thresholds.shape[0]
    plt.plot(thresholds, precisions[0:threshold_boundary], linestyle='--', label='precision')
    plt.plot(thresholds, recalls[0:threshold_boundary],label='recall')
    
    # XÏ∂ï(threshold)Ïùò ScaleÏùÑ 0 ~ 1 Îã®ÏúÑÎ°ú Î≥ÄÍ≤Ω
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1),2))
    
    plt.xlabel('Threshold value'); plt.ylabel('Precision and Recall value')
    plt.legend()
    plt.grid()
    plt.show()
```

### ROC Curve
- ÌòÑÏû¨ Î™®Îç∏Ïùò ROC CurveÎ•º ÏãúÍ∞ÅÌôî

```python
from sklearn.metrics import roc_curve

# ROC CurveÎ•º ÏãúÍ∞ÅÌôîÌï¥Ï£ºÎäî Ìï®Ïàò
def roc_curve_plot(y_test , proba_class1):
#     ÏûÑÍ≥ÑÍ∞íÏóê Îî∞Î•∏ FPR, TPR Í∞í
    fprs, tprs, thresholds = roc_curve(y_test, proba_class1)

#     ROC CurveÎ•º plot Í≥°ÏÑ†ÏúºÎ°ú Í∑∏Î¶º. 
    plt.plot(fprs , tprs, label='ROC')
#     Í∞ÄÏö¥Îç∞ ÎåÄÍ∞ÅÏÑ† ÏßÅÏÑ†ÏùÑ Í∑∏Î¶º. 
#     TPRÍ≥º FPRÏù¥ ÎèôÏùºÌïú ÎπÑÏú®Î°ú Îñ®Ïñ¥ÏßÑÎã§Îäî Í≤ÉÏùÄ Î™®Îç∏Ïù¥ ÏñëÏÑ±Í≥º ÏùåÏÑ±ÏùÑ Íµ¨Î≥ÑÌïòÏßÄ Î™ªÌïúÎã§Îäî Í≤ÉÏùÑ ÏùòÎØ∏ÌïúÎã§.
#     Îã§Î•∏ Î∂ÑÎ•òÍ∏∞Î•º ÌåêÎã®ÌïòÍ∏∞ ÏúÑÌïú Í∏∞Ï§ÄÏÑ†ÏúºÎ°ú ÏÇ¨Ïö©ÎêòÎ©∞, 
#     ÎåÄÍ∞ÅÏÑ†Ïóê Í∞ÄÍπåÏö∏ ÏàòÎ°ù ÏòàÏ∏°Ïóê ÏÇ¨Ïö©ÌïòÍ∏∞ ÌûòÎì† Î™®Îç∏Ïù¥ÎùºÎäî ÎúªÏù¥Îã§.
    plt.plot([0, 1], [0, 1], 'k--', label='Standard')
    
    # XÏ∂ï(FPR)Ïùò ScaleÏùÑ 0.1 Îã®ÏúÑÎ°ú Î≥ÄÍ≤Ω
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1),2))
    plt.xlim(0,1); plt.ylim(0,1)
    plt.xlabel('FPR( 1 - Sensitivity )'); plt.ylabel('TPR( Recall )')
    plt.legend()
    plt.show()
```

```python
# ÏúÑ Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ROC CurveÎ•º ÏãúÍ∞ÅÌôî
roc_curve_plot(y_test, dtc.predict_proba(X_test)[:, 1].reshape(-1, 1))
```

  Ìï¥Îãπ ÏàòÏπòÎì§ Î™®Îëê 0.9 Ïù¥ÏÉÅÏùò ÎÜíÏùÄ ÏàòÏπòÎ•º Î≥¥Ïù¥Îäî Í≤ÉÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏóàÏäµÎãàÎã§.
- Ïù¥Îäî Ï†ÑÏ≤¥Ï†ÅÏù∏ Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÍ∞Ä Î∂ÄÏ°±ÌñàÏúºÎ©∞, targetÏùò Îç∞Ïù¥ÌÑ∞(Ture/False) ÎπÑÏ§ëÏù¥ ÌÅ∞ Ï∞®Ïù¥Î•º Î≥¥ÏòÄÍ∏∞Ïóê Î∞úÏÉùÌïú ÌòÑÏÉÅÏûÖÎãàÎã§.
- Îî∞ÎùºÏÑú Ïù¥ÌõÑ Ï∂îÍ∞ÄÏ†ÅÏù∏ Îç∞Ïù¥ÌÑ∞ ÏàòÏßë Î∞è Ï†ÑÏ≤òÎ¶¨ Î∞©ÏãùÏóê ÎåÄÌïú ÌïôÏäµÏù¥ ÌïÑÏöîÌï† Í≤ÉÏúºÎ°ú Î≥¥ÏûÖÎãàÎã§.
