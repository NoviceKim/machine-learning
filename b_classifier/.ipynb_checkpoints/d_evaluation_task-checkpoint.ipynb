### Confusion Matrix Task

##### 가정: 고암(Abnormal) 환자와 정상(Normal) 환자를 분류하는 의료 진단 시나리오.

- 양성(Positive): 고암 환자를 의미.
- 음성(Negative): 정상 환자를 의미.

#### 문제 1: 유방암 진단

>True Positive (TP): 모델이 고암이라고 예측한 환자이며, 실제로도 고암인 환자 (맞힘)  
False Positive (FP): 모델이 고암이라고 예측한 환자이지만, 실제로는 정상인 환자 (틀림, error I)  
True Negative (TN): 모델이 정상이라고 예측했으며, 실제로도 정상이었던 환자 (맞힘)  
False Negative (FN): 모델이 정상이라고 예측했지만, 실제로도 정상이었던 환자 (틀림, error II)  

>정밀도 (Precision): 모델이 고암이라고 예측 한 환자들 중, 실제로도 고암인 환자의 비율 (예측 → 정답)  
재현율 (Recall): 실제 고암 환자 중에서 모델이 놓치지 않고 잡아낸 고암 환자의 비율 (정답 → 예측)  

> 📌 이 경우, FN의 비율을 낮춰야 하기 때문에 전반적인 Negative(정상) 예측 비율을 낮출 필요가 있다.  
따라서, 임계값(Threshold Value)을 낮춤으로서 재현율을 향상시켜야 한다.

##### 가정: 정상 메일과 스팸 메일을 분류하는 스팸 메일 분류 시나리오.

- 양성(Positive): 스팸 메일을 의미.
- 음성(Negative): 정상 메일을 의미.

#### 문제 2: 스팸 메일 필터링

> True Positive (TP): 모델이 스팸 메일이라고 예측했으며, 실제로도 스팸 메일인 경우 (맞힘)  
False Positive (FP): 모델이 스팸 메일이라고 예측했지만, 실제로는 정상 메일이었던 경우 (틀림, error I)  
True Negative (TN): 모델이 정상 메일이라고 예측했으며, 실제로도 정상 메일이었던 경우 (맞힘)  
False Negative (FN): 모델이 정상 메일이라고 예측했지만, 실제로는 스팸 메일이었던 경우 (틀림, error II)  

> 정밀도 (Precision): 모델이 스팸 메일이라고 예측한 메일 중 실제로도 스팸이었던 메일의 비율  
재현율 (Recall): 실제 스팸 메일 중에서 모델이 놓치지 않고 잡아낸 스팸 메일의 비율  

> 📌 이 경우, FP의 비율을 낮춰야 하기 때문에 전반적인 Positive(스팸 메일) 예측 비율을 낮출 필요가 있다.  
따라서 임계값을 높임으로서 정밀도를 향상시켜야 한다.  

##### 가정: 범죄 예측 시나리오.

- 양성(Positive): 범죄가 예측되는 경우 (예: 범죄가 발생함).
- 음성(Negative): 범죄가 예측되지 않는 경우 (예: 범죄가 발생하지 않음).

#### 문제 3: 범죄 예측

> True Positive (TP): 모델이 범죄가 예측된다고 했으며, 실제로도 범죄가 발생한 경우 (맞힘)  
False Positive (FP): 모델이 범죄가 예측된다고 했지만, 실제로는 범죄가 발생하지 않은 경우 (틀림, error I)  
True Negative (TN): 모델이 범죄가 예측되지 않는다고 예측했으며, 실제로도 범죄가 발생하지 않은 경우 (맞힘)  
False Negative (FN): 모델이 범죄가 예측되지 않는다고 예측했지만, 실제로는 범죄가 발생한 경우 (틀림, error II)  

> 정밀도 (Precision): 모델이 범죄가 일어날 것이라고 예측한 건 중, 실제로도 범죄가 일어난 건의 비율  
재현율 (Recall): 실제 범죄 중에서 모델이 놓치지 않고 범죄를 예측한 건의 비율  

> 📌 이 경우, FN의 비율을 낮춰야 하기 때문에 전반적인 Negative(범죄 발생 안 함) 예측 비율을 낮출 필요가 있다.  
따라서, 임계값을 낮춤으로서 재현율을 올려야 한다.  

##### 가정: 검색 엔진에서 웹 페이지의 검색 결과를 분류하는 검색 품질 모델 서비스 시나리오.

- 양성(Positive): 검색 결과가 사용자의 의도에 부합하는 경우 (예: 정확한 검색 결과).
- 음성(Negative): 검색 결과가 사용자의 의도에 부합하지 않는 경우 (예: 부정확한 또는 불량한 검색 결과).

#### 문제 4: 검색 품질 모델

> True Positive (TP): 모델이 정확하다고 예측했고, 실제로도 결과가 정확한 경우 (맞힘)  
False Positive (FP): 모델이 정확하다고 예측했지만, 실제로는 결과가 부정확한 경우 (틀림, error I)  
True Negative (TN): 모델이 부정확하다고 예측했으며, 실제로도 결과가 부정확한 경우 (맞힘)  
False Negative (FN): 모델이 부정확하다고 예측했지만, 실제로는 결과가 정확한 경우 (틀림, error II)  

> 정밀도 (Precision): 모델이 정확하다고 예측한 검색 결과 중 실제로도 정확했던 건의 비율  
재현율 (Recall): 실제로 정확한 검색 결과 중에서 모델이 놓치지 않고 정확하다고 예측한 건의 비율  

> 📌 이 경우, FP의 비율을 낮춰야 하기 때문에 전반적인 Positive(정확한 결과) 예측 비율을 낮출 필요가 있다.  
따라서, 임계값을 높임으로서 정밀도를 올려야 한다.  

##### 가정: 보안 시스템에서 침입 감지 시스템 시나리오.

- 양성(Positive): 침입이 감지된 경우 (예: 보안 이벤트가 발생함).
- 음성(Negative): 침입이 감지되지 않은 경우 (예: 정상 상황).  

#### 문제 5: 침입 감지 시스템

> True Positive (TP): 모델이 침입이 감지됐다고 예측했으며, 실제로도 침입이 감지된 경우 (맞힘)  
False Positive (FP): 모델이 침입이 감지됐다고 예측했지만, 실제로는 침입이 발생하지 않은 경우 (틀림, error I)  
True Negative (TN): 모델이 침입이 감지되지 않는다고 했으며, 실제로도 침입이 감지되지 않은 경우 (맞힘)  
False Negative (FN): 모델이 침입이 감지되지 않는다고 했지만, 실제로는 침입이 감지된 경우 (틀림, error 2)  

> 정밀도 (Precision): 모델이 침입이 감지됐다고 예측한 건 중, 실제로도 침입이 감지된 건의 비율  
재현율 (Recall): 실제 침입 중에서 모델이 놓치지 않고 잡아낸 침입 발생 건의 비율  

> 📌 이 경우, FN의 비율을 낮춰야 하기 때문에 전반적인 Negative(정상) 예측 비율을 낮출 필요가 있다.  
따라서, 임계값을 낮춤으로서 재현율을 향상시켜야 한다.  

>  단, 일반 가정에서 해당 모델을 채용하는 경우 침입이 발생했다고 판단하는 비율이 높아져서  
정상적인 상황에서도 위험 감지 신호를 보내게 되면 사용자 입장에서 매우 불안해질 수 있다.  
따라서 이런 상황에서는 FP의 비율을 낮춰서 정밀도를 향상시키는 것도 고려해 볼 수 있다.

### ⚠️ 요점 정리
- 정밀도
> - 모델이 예측한 답안을 먼저 보고 실제 정답과 비교  
> - 임계값을 높임으로서 Positive라고 예측한 비율을 낮춘다.  
> - TP이 비중이 상승하고, FP의 비중은 낮아진다.  

- 재현율
> - 실제 정답을 먼저 보고 모델이 예측한 답안과 비교  
> - 임계값을 낮춤으로서 Negative라고 예측한 비율을 낮춘다.  
> - TP의 비중이 상승하고, FN의 비중은 낮아진다.
