### 데이터 세트: 중고차 가격 회귀 분석

---

#### Features (총 12개)
- Car_ID: 자동차 별 고유 ID
- Brand: 브랜드
- Model: 모델명
- Year: 생산 년도
- Kilometers_Driven: 총 주행 거리
- Fuel_Type: 연료 유형
> - Petrol: 휘발유
> - Diesel: 경유

- Transmission: 기어 유형
> - Manual: 수동
> - Automatic: 자동

- Owner_Type: 자동차의 이전 소유주 수 (해당 중고차가 중고차 시장에 몇 번째 나온 것인가를 의미)
> - First(1명) ~ Third(3명)

- Mileage: 연비 (km/l)
- Engine: 엔진 배기량 (cc)
- Power: 최대 제동 마력 (bhp)
- Seats: 좌석 수
> - 4 ~ 7

#### Targets
- Price: 위 feature들로 예측한 중고차의 가격

```python
import pandas as pd
import numpy as np

c_df = pd.read_csv('../../datasets/car_price.csv')
c_df
```

```python
# 데이터 세트 정보 확인
c_df.info()
```

### 데이터 전처리

```python
# 결측치 검사
c_df.isna().sum()
```

```python
# 중복 검사
c_df.duplicated().sum()
```

```python
# 불필요한 feature를 제거한 전처리용 데이터 세트 생성
pre_c_df = c_df.drop(labels=['Car_ID', 'Brand', 'Model'], axis=1)
pre_c_df
```

#### 레이블 인코딩
- Fuel_Type과 Transmission은 클래스가 2가지이고,  
  Owner_Type은 클래스 간 순서가 있기 때문에 이 feature들에 전부 레이블 인코더 적용

```python
from sklearn.preprocessing import LabelEncoder

# columns에 담긴 feature들에 LabelEncoder 적용
columns = ['Fuel_Type', 'Transmission', 'Owner_Type']
encoders = {}

for column in columns:
    encoder = LabelEncoder()
    result = encoder.fit_transform(pre_c_df[column])
    pre_c_df[column] = result
    encoders[column] = encoder.classes_

print(encoders)
```

```python
# 레이블 인코딩 후의 데이터 세트 확인
pre_c_df
```

### 1st Cycle - 선형 회귀 모델 학습

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 데이터 세트 분할 후 선형 회귀 모델로 훈련
features, targets = pre_c_df.iloc[:, :-1], pre_c_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

# 선형 회귀 모델의 평가 지표(R2 점수 포함)를 출력해주는 함수
def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, MSLE, RMSLE, R2))
```

```python
# 테스트 데이터 예측 후 평가
prediction = l_r.predict(X_test)

get_evaluation(y_test, prediction)
```

#### 과적합 발생 여부 확인
- train 데이터 세트 만들고 다시 분할해서 validation 데이터 세트 생성
- validation 데이터 세트를 선형 회귀 모델로 학습한 뒤, test 데이터의 R2와 비교

```python
# X_train과 y_train 병합해서 train 데이터 세트 생성
train_c_df = pd.concat([X_train, y_train], axis=1).reset_index(drop=True)
train_c_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# train 데이터 세트 분할 후 선형 회귀 모델로 훈련
validation_features, validation_targets = train_c_df.iloc[:, :-1], train_c_df.iloc[:, -1]

X_va_train, X_va_test, y_va_train, y_va_test = \
train_test_split(validation_features, validation_targets, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_va_train, y_va_train)
```

```python
# 테스트 데이터 예측 후 평가
prediction = l_r.predict(X_va_test)

get_evaluation(y_va_test, prediction)
```

#### 1st Cycle - 과적합 발생 여부 확인 결과
- validation 데이터 세트로 학습한 결과, test 데이터 세트에 비해 R2가 크게 감소함
- 이로 인해 현재 모델에 과적합이 발생했다고 판단할 수 있음
