### Dimension Reduction Task01
- 뇌암 유전자 발현 데이터 세트를 회귀 분석하되, 차원을 축소하여 진행해본다.

```python
import pandas as pd
import numpy as np

b_df = pd.read_csv('./datasets/brain_cancer.csv', low_memory=False)
b_df
```

```python
# 총 결측치 수 확인
b_df.isna().sum().sum()
```

```python
# 데이터 세트 정보 출력
b_df.info()
```

### 데이터 전처리 

```python
# 타겟 데이터(type)의 value counts 출력
b_df['type'].value_counts()
```

#### 중간 분석
- samples는 불필요하므로 제거
- 타겟 데이터(type)는 일단 레이블 인코딩하고, 추후 원핫 인코딩으로 재시도

```python
# 불필요한 컬럼 제거
pre_b_df = b_df.drop(labels='samples', axis=1)
pre_b_df
```

#### 레이블 인코딩

```python
from sklearn.preprocessing import LabelEncoder

# 타겟 데이터에 레이블 인코더 적용
label_encoder = LabelEncoder()
result = label_encoder.fit_transform(pre_b_df['type'])

# 레이블 인코더 적용 후 결과와 원래 데이터 클래스 출력
print(result)
print(label_encoder.classes_)

# 레이블 인코더가 적용된 후의 값을 타겟 컬럼에 적용
pre_b_df['type'] = result

# 데이터 세트 확인
display(pre_b_df)
```

#### 표준화 후 이상치 제거

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()

# feature 데이터 값 표준화
result = std.fit_transform(pre_b_df.iloc[:, 1:])

std_pre_b_df = pd.DataFrame(result, columns=pre_b_df.iloc[:, 1:].columns)
std_pre_b_df
```

```python
# 각 feature 별로 -1.96 ~ 1.96 범위를 벗어나는 데이터 제거
for column in std_pre_b_df.columns:
    std_pre_b_df = std_pre_b_df[std_pre_b_df[column].between(-1.96, 1.96)]

std_pre_b_df
```

#### 이슈 발생
- 표준화 후 이상치를 제거할 경우 남는 데이터가 없음
- 따라서 이상치 제거는 생략

```python
pre_b_df
```

```python
pre_b_df['type'].value_counts()
```

#### ✏️ 메모
- 만약 레이블 인코딩 이후의 과정에서  
**ValueError: Unknown label type: 'unknown'**  
  에러 발생 시, 타겟 데이터의 dtype 확인
- 만약 dtype이 object('O') 라면, 타겟 데이터에 .apply(pd.to_numeric)를 사용하여 숫자형으로 변환하는 것이 하나의 방법

### 데이터 세트 분할

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# 데이터 세트 분할
features, targets = pre_b_df.iloc[:, 1:], pre_b_df.iloc[:, 0]

X_train, X_test, y_train, y_test = train_test_split(features, targets, stratify=targets, test_size=0.2, random_state=124)

# 오버 샘플링
smote = SMOTE(random_state=124)

over_X_train, over_y_train = smote.fit_resample(X_train, y_train)
```

```python
# 훈련 데이터와 테스트 데이터를 각각 병합하여 새로운 데이터 세트 생성
train_df = pd.concat([over_X_train, over_y_train], axis=1).reset_index(drop=True)
test_df = pd.concat([X_test, y_test], axis=1).reset_index(drop=True)
```

```python
# 훈련용, 테스트용 데이터 세트 확인
display(train_df)
```

```python
display(test_df)
```

### PCA 방식으로 차원 축소

```python
from sklearn.decomposition import PCA

# PCA 방식으로 데이터 세트 차원 축소
pca = PCA(n_components=2)

pca_train = pca.fit_transform(train_df.iloc[:, :-1])
pca_test = pca.fit_transform(test_df.iloc[:, :-1])
```

```python
# 차원 축소 여부 확인
print(pca_train.shape, pca_test.shape)
```

```python
# 훈련용 / 테스트용 데이터 세트를 차원 축소한 새로운 데이터 세트 생성
pca_columns = [f'pca{i + 1}' for i in range(pca_train.shape[1])]
pca_train_df = pd.DataFrame(pca_train, columns=pca_columns)
pca_train_df.loc[:, 'target'] = train_df['type']

pca_columns = [f'pca{i + 1}' for i in range(pca_test.shape[1])]
pca_test_df = pd.DataFrame(pca_test, columns=pca_columns)
pca_test_df.loc[:, 'target'] = test_df['type']
```

```python
# 차원 축소된 데이터 세트 확인
display(pca_train_df)
display(pca_test_df)
```

```python
# 기존 데이터에 대한 보존률 확인 - 현재 약 34% 보존됨
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 현재 데이터 분포를 산점도로 시각화
sns.scatterplot(x='pca1', y='pca2', hue='target', data=pca_train_df, alpha=1)
plt.show()
```

```python
# 기존 데이터에 대한 보존률 확인 - 현재 약 60% 보존됨
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
```

- 축소 후 차원을 5차원으로 늘려본 결과, 기존 데이터 대비 약 60%의 보존률을 보임
- 이전 시도보다는 향상되었으나 여전히 높지 않은 수치이기에 차원을 더 늘릴 필요가 있음

```python
from sklearn.decomposition import PCA

# 축소 후 차원을 8차원으로 조정
pca = PCA(n_components=8)

pca_train = pca.fit_transform(train_df.iloc[:, :-1])
pca_test = pca.fit_transform(test_df.iloc[:, :-1])
```

```python
# 차원 축소 여부 확인
print(pca_train.shape, pca_test.shape)
```

```python
# 훈련용 / 테스트용 데이터 세트를 차원 축소한 새로운 데이터 세트 생성
pca_columns = [f'pca{i + 1}' for i in range(pca_train.shape[1])]
pca_train_df = pd.DataFrame(pca_train, columns=pca_columns)
pca_train_df.loc[:, 'target'] = train_df['type']

pca_columns = [f'pca{i + 1}' for i in range(pca_test.shape[1])]
pca_test_df = pd.DataFrame(pca_test, columns=pca_columns)
pca_test_df.loc[:, 'target'] = test_df['type']
```

```python
# 차원 축소된 데이터 세트 확인
display(pca_train_df)
display(pca_test_df)
```

```python
# 기존 데이터에 대한 보존률 확인 - 현재 약 72% 보존됨
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
```

#### 중간 분석
- 축소 후 차원을 8차원으로 늘려본 결과, 기존 데이터 대비 약 72%의 보존률을 보임
- 타겟 데이터가 범주형이기에 분류 모델로 학습 진행

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

features, targets = pca_train_df.iloc[:, :-1], pca_train_df.iloc[:, -1]
parameters = {'max_depth': [5, 10, 15, 20], 'min_samples_split': [10, 50, 100]}

rfc = RandomForestClassifier()

# Random Forest 분류 모델 교차 검증
grid_rfc = GridSearchCV(rfc, param_grid=parameters, cv=5, return_train_score=True, scoring='accuracy')
grid_rfc.fit(features, targets)
```

```python
# 하이퍼 파라미터 조합 별 평가 점수를 데이터프레임으로 만들어 출력
result_df = pd.DataFrame(grid_rfc.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

def get_evaluation(y_test, prediction, classifier=None, X_test=None):
    confusion = confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test , prediction)
    precision = precision_score(y_test , prediction, average='macro')
    recall = recall_score(y_test , prediction, average='macro')
    f1 = f1_score(y_test, prediction, average='macro')
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}'.format(accuracy, precision, recall, f1))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
# 테스트 데이터 세트 분리
X_test, y_test = pca_test_df.iloc[:, :-1], pca_test_df.iloc[:, -1]

# 최적의 하이퍼 파라미터로 튜닝 후 오차 행렬 시각화
rfc = grid_rfc.best_estimator_
prediction = rfc.predict(X_test)
get_evaluation(y_test, prediction, rfc, X_test)
```

#### PCA 방식으로 차원 축소 후의 학습 결과
- Random Forest 모델로 교차 검증 후 하이퍼 파라미터 최적화까지 완료했음에도 평가 지표가 0.3 정도의 낮은 수치를 보임
- 이는 차원 축소 과정에서 대다수의 feature가 삭제되었으며, 표준화도 진행하지 않았기 때문인 것으로 추정

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# 데이터 세트 분할
features, targets = pre_b_df.iloc[:, 1:], pre_b_df.iloc[:, 0]

X_train, X_test, y_train, y_test = train_test_split(features, targets, stratify=targets, test_size=0.2, random_state=124)

# 오버 샘플링
smote = SMOTE(random_state=124)

over_X_train, over_y_train = smote.fit_resample(X_train, y_train)
```

```python
# 훈련 데이터와 테스트 데이터를 각각 병합하여 새로운 데이터 세트 생성
train_df = pd.concat([over_X_train, over_y_train], axis=1).reset_index(drop=True)
test_df = pd.concat([X_test, y_test], axis=1).reset_index(drop=True)
```

```python
from sklearn.decomposition import PCA

for i in range(14):
    pca = PCA(n_components=(i + 1))

    pca_train = pca.fit_transform(over_X_train)
    
    # 차원 축소 후 데이터 보존률 확인
    print(pca.explained_variance_ratio_.sum())
```

```python
from sklearn.decomposition import PCA

# 축소 후 차원을 2차원으로 조정 - 산점도로 시각화하기 위함
pca = PCA(n_components=2)

# PCA 방식을 통해 기존 feature들을 2차원으로 축소
pca_train = pca.fit_transform(over_X_train)

# 차원 축소가 완료된 훈련용 데이터 세트 생성
pca_columns = [f'pca{i + 1}' for i in range(pca_train.shape[1])]
pca_train_df = pd.DataFrame(pca_train, columns=pca_columns)
pca_train_df.loc[:, 'target'] = over_y_train
```

#### 중간 분석
- PCA 방식을 차원 축소를 했을 때, 위 산점도와 같이 각 클래스 별 영역을 구분하기 어려움
- 이는 모델의 분류 성능을 필요 이상으로 저하시킬 수 있기 때문에, LDA 방식으로 차원 축소 실행

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# LDA 방식 적용 후 차원을 1씩 늘려가면서 각 차원 별 보존률 출력
# LDA 방식으로 차원 축소 시, 최대 차원은 타겟 데이터의 클래스 수 - 1(현재 타겟의 클래스 수가 5이므로 최대 4차원까지만 가능)
for i in range(4):
    lda = LinearDiscriminantAnalysis(n_components=(i + 1))
    
    lda_train = lda.fit_transform(over_X_train, over_y_train)

    print(lda.explained_variance_ratio_.sum())
```

#### 중간 분석
- 2차원으로 축소했을 때, 보존률이 0.76으로 적절한 결과를 보임
- 따라서 아래 과정 역시 2차원으로 축소한 다음 실행

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# LDA 방식을 통해 2차원으로 축소
lda = LinearDiscriminantAnalysis(n_components=2)

# LDA 방식으로 차원 축소 시, 타겟 데이터를 같이 전달한다
lda_train = lda.fit_transform(over_X_train, over_y_train)

lda_columns = [f'lda{i + 1}' for i in range(lda_train.shape[1])]
lda_train_df = pd.DataFrame(lda_train, columns=lda_columns)
lda_train_df.loc[:, 'target'] = over_y_train
```

  PCA 방식으로 축소했을 때에 비해 클래스 간 영역 구분이 보다 확실한 모습을 보임

#### 파이프라인 구축
- 표준화 → LDA(2차원) → Random Froest 분류 모델을 순서대로 거치는 파이프라인 구축
- 설정한 하이퍼 파라미터로 교차 검증

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

# 교차 검증에 사용할 하이퍼 파라미터와 Random Forest 분류 모델 객체
parameters = {'rfc__max_depth': [5, 10, 20], 'rfc__min_samples_split': [10, 50, 100]}
rfc = RandomForestClassifier()

# 파이프라인 구축
rfc_pipe = Pipeline([('std', StandardScaler()),
                     ('lda', LinearDiscriminantAnalysis(n_components=2)),
                     ('rfc', rfc)])

# 위의 파이프라인으로 훈련 데이터 교차 검증
g_rfc = GridSearchCV(rfc_pipe, param_grid=parameters, cv=3, scoring='accuracy')
g_rfc.fit(over_X_train, over_y_train)
```

```python
# 각 하이퍼 파라미터 별 평가 점수를 데이터 세트 형식으로 출력
result_df = pd.DataFrame(g_rfc.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

# 다중 분류 모델의 평가 지표를 출력하고, 분류기를 전달 받으면 추가로 오차 행렬 시각화도 해주는 함수
# SMOTE로 타겟 데이터의 비중을 균일하게 맞췄기 때문에 macro 방식으로 평균 산출 
def get_evaluation(y_test, prediction, classifier=None, X_test=None):
    # 다중 분류 모델의 평가 지표 출력
    confusion = confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test , prediction)
    precision = precision_score(y_test , prediction, average='macro')
    recall = recall_score(y_test , prediction, average='macro')
    f1 = f1_score(y_test, prediction, average='macro')
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}'.format(accuracy, precision, recall, f1))
    print("#" * 80)

    # 분류기와 테스트용 feature(문제)를 전달 받으면 오차 행렬 시각화
    if classifier is not None and X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
# 최적의 하이퍼 파라미터로 튜닝 후 테스트 데이터 예측
rfc = g_rfc.best_estimator_
prediction = rfc.predict(X_test)

# 테스트 데이터의 예측을 평가
get_evaluation(y_test, prediction)
```
